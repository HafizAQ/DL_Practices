```python
# Creating Graph using networkx

import networkx as nx
import matplotlib.pyplot as plt

#creating a graph
graph = nx.Graph()

# Edges to the graph
graph.add_edge('A', 'B')
graph.add_edge('A', 'D')
graph.add_edge('D', 'C')
graph.add_edge('B', 'D')

#Rendering/ Showing the graph
plt.figure(figsize=(10,4))
nx.draw(graph, with_labels=True)
plt.show()
```
